```python
import akshare as ak
import pandas as pd
import vectorbt as vbt
```

```python
stocks = ak.index_stock_cons_csindex(symbol="000300")['成分券代码'].tolist()[:50]
prices = pd.DataFrame()
```

```python
for code in stocks:
    df = ak.stock_zh_a_hist(symbol=code, adjust="qfq")['收盘']
    prices[code] = df
    print(code)
```

```python
# 简单因子：过去20日动量
momentum = prices.pct_change(20).iloc[-1]  # 最新排名
signal = momentum.rank(pct=True) > 0.8  # Top 20%买

pf = vbt.Portfolio.from_signals(prices, entries=signal.reindex(prices.index), freq='1M')  # 月调
print(pf.stats())  # 年化/回撤一键出
```
